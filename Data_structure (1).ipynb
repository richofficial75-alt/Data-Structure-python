## ***Data Types and Structures Questions***

**1. What are data structures, and why are they important?**

 - A data structure is a way of organizing and storing data in a computer so that it can be accessed and modified efficiently.

 **2.  Explain the difference between mutable and immutable data types with examples.**

 - Mutable data types can be changed or modified after creation — you can alter their contents without creating a new object in memory and Immutable data types cannot be changed after creation.
 If you try to modify them, a new object is created in memory.

 **3. What are the main differences between lists and tuples in Python?**

  -  A mutable (changeable) ordered collection of items  and an immutable (unchangeable) ordered collection of items.

  **4. Describe how dictionaries store data?**

  - In Python, dictionaries store data in the form of key–value pairs, allowing for fast and efficient data retrieval.

  **5. Why might you use a set instead of a list in Python?**

  - We might use a set instead of a list in Python when you need to store unique items and perform fast membership testing.

  **6. What is a string in Python, and how is it different from a list?**

  - A string is an immutable sequence of Unicode characters.
  - A string is an immutable sequence of characters, while a list is a mutable collection that can hold any type of data.

  **7. How do tuples ensure data integrity in Python?**

  - Tuples ensure data integrity in Python primarily through their immutability — meaning once a tuple is created, its contents cannot be changed.

  **8.What is a hash table, and how does it relate to dictionaries in Python?**

  - A hash table is a data structure that stores data in key–value pairs and allows very fast access, insertion, and deletion using a process called hashing.
  - In Python, the dictionary (dict) is implemented internally using a hash table.

  **9. Can lists contain different data types in Python?**

  - Yes!
In Python, lists can contain elements of different data types — including numbers, strings, booleans, other lists, and even objects.

 **10.  Explain why strings are immutable in Python**

 - Strings in Python are immutable, meaning they cannot be changed after creation.
Once you create a string, you cannot modify its characters, insert, or delete anything within it — any operation that seems to “change” a string actually creates a new one.

**11. What advantages do dictionaries offer over lists for certain tasks?**

 - Dictionaries in Python offer several advantages over lists, especially when you need to associate data with meaningful keys or retrieve items quickly.

 **12. Describe a scenario where using a tuple would be preferable over a list.**

 - A tuple is preferable over a list when you need to store a fixed collection of items that should not change during the program’s execution — ensuring data integrity and faster performance.

 **13. How do sets handle duplicate values in Python?**

 - When we create a set, Python internally uses a hash table to store its elements.
 Each element’s hash value is used to determine its position in memory.

 **14. How does the “in” keyword work differently for lists and dictionaries?**

 - The in keyword in Python is used to check membership, but it works differently for lists and dictionaries.

 **15. Can you modify the elements of a tuple? Explain why or why not?**

 - No, you cannot modify the elements of a tuple in Python.
 Tuples are immutable, meaning once they are created, their elements cannot be changed, added, or removed.

 **16. What is a nested dictionary, and give an example of its use case?**

 - A nested dictionary in Python is a dictionary inside another dictionary — it allows you to store and organize complex, hierarchical data.

 **17.  Describe the time complexity of accessing elements in a dictionary.**

 - The time complexity of accessing elements in a Python dictionary is typically:
 Average case: O(1) and Worst case: O(n)

 **18. In what situations are lists preferred over dictionaries?**

 - Lists are preferred over dictionaries when you need to store an ordered collection of items and don’t need key–value pairs.
- Lists preserve the order of elements, so you can   access them by position (index).

**19. Why are dictionaries considered unordered, and how does that affect data retrieval?**

 - Dictionaries in Python are often described as unordered collections because their elements are stored based on hash values of keys, not on the order in which they were added (at least conceptually).

 **20. Explain the difference between a list and a dictionary in terms of data retrieval.**

 - The main difference between a list and a dictionary in Python lies in how you retrieve data — i.e., how you access stored elements.

**1. Write a code to create a string with your name and print it.**

**2. Write a code to find the length of the string "Hello World".**

```python
# Create a string
text = "Hello World"

# Find the length of the string
length = len(text)

# Print the length
print(length)
```

**3. Write a code to slice the first 3 characters from the string "Python Programming".**

```python
# Create a string
text = "Python Programming"

# Slice the first 3 characters
first_three = text[:3]

# Print the result
print(first_three)
```

**4. Write a code to convert the string "hello" to uppercase.**

```python
# Create a string
text = "hello"

# Convert to uppercase
uppercase_text = text.upper()

# Print the result
print(uppercase_text)
```

**5. Write a code to replace the word "apple" with "orange" in the string "I like apple.**

```python
# Create a string
text = "I like apple"

# Replace 'apple' with 'orange'
new_text = text.replace("apple", "orange")

# Print the result
print(new_text)
```

7. Write a code to append the number 10 to the list [1, 2, 3, 4]

```python
# Create a list
numbers = [1, 2, 3, 4]

# Append 10 to the list
numbers.append(10)

# Print the updated list
print(numbers)
```

**8. Write a code to remove the number 3 from the list [1, 2, 3, 4, 5]**

**9. Write a code to access the second element in the list ['a', 'b', 'c', 'd']**

```python
# Create a list
letters = ['a', 'b', 'c', 'd']

# Access the second element (index 1)
second_element = letters[1]

# Print the result
print(second_element)
```

**10. Write a code to reverse the list [10, 20, 30, 40, 50].**
